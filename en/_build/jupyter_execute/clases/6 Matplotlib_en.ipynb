We define a set of 51 equate data between 0 and 4, and the function

$$
y = f (x) = 2.5 E^{-1.3 x} + 0.5. E^{-1.6 x}
$$

In addition, we define a data set _With noise_, adding to $ and $ a gaussian noise on average 0 and dispersion 1 (see `help (np.random.normal)`):

```python
x = np.linspace(0,4,51)
y = 2.5 * np.exp(-1.3 * x) + 0.5 * np.exp(-1.6 * x)
ruido = 0.2 * np.random.normal(size=x.size)
medicion = y + ruido

print(x.size)
```

### Axes and legends names

We are now adding names to the axes and curves.

To add names to the curves, we have to add a `Label`, in this case in the same` plot () `command, and then
Show it with `legend ()

```python
plt.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
plt.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
plt.legend(title='DATOS')
plt.show()
```

To add names to the axes we use `xlabel` and` ylabel`:
The titles to the figure can be added with `Title`:

```python
plt.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
plt.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
plt.legend(title='DATOS')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Valor [mV]')
plt.title('RESULTADOS')
plt.show()
```

We can also graph vertical and horizontal lines using `axvline` and` axhline`

```python
plt.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
plt.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
plt.legend(title='DATOS')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Valor [mV]')
plt.title('RESULTADOS')
plt.axvline(x=2.5, color='black')
plt.axhline(y = 1.5, color='black')
plt.show()
```

To pass on logarithmic scale we use `xscale` or` yscale`

```python
plt.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
plt.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
plt.legend(title='DATOS')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Valor [mV]')
plt.title('RESULTADOS')
# plt.xscale('log')
plt.yscale('log')
plt.grid(color='green', linestyle='dashed', linewidth=1)
plt.show()
```

### Two graphics in the same figure

There are several functions that allow to put more than one graph in the same figure. Let's look at an example using `Figure` and` add_sublot`

```python
fig_1 = plt.figure(1, figsize=[20, 5])
graf1 = fig_1.add_subplot(121) # 1: number of rows 2: number of columns 1: the number of subploot
graf2 = fig_1.add_subplot(122) # 1: number of rows 2: number of columns 2: the number of subploot

graf1.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
graf1.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
graf1.legend(title='DATOS')
graf1.grid(color='green', linestyle='dashed', linewidth=1)

graf1.set_xlabel('Tiempo [seg]', fontsize='x-large')
graf1.set_ylabel('Valor [mV]', fontsize='x-large')
graf1.set_title('RESULTADOS ESCALA LOG')
graf1.set_xscale('log')

graf2.plot(x,y,linewidth=5, linestyle='dashed', label='Teora')
graf2.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=4, label='Medicin')
graf2.legend(title='DATOS')
graf2.grid(color='green', linestyle='dashed', linewidth=1)

graf2.set_xlabel('Tiempo [seg]')
graf2.set_ylabel('Valor [mV]')
graf2.set_title('RESULTADOS')
plt.show()
```

Here can also be used latex type format for part of the text. If we use an expression enclosed among the `$` symbols, Matpletlib interprets that it is written in (a subset) of Latex.

** Matpletlib ** has an internal symbols processor to show the latex notation that recognizes the most common elements, or it can be chosen to use an external latex processor.

```python
plt.plot(x,y,linewidth=5, linestyle='dashed', label=r"$\frac{\sin(e^2)}{[2 + \cos (x)]^2}$")
plt.plot(x,medicion, '-o', color='red', markersize=8, markeredgecolor='black',markevery=2, label='Medicin')
plt.legend(title='DATOS', fontsize='x-large', shadow=False, loc=3)
plt.xlabel('Tiempo [seg]')
plt.ylabel('Valor [mV]')
plt.title('RESULTADOS: $f(x)$')
plt.xscale('log')
plt.grid(color='green', linestyle='dashed', linewidth=1)
plt.show()
```

## Reading an image

In addition to performing 2D graphics, `Matpletlib` allows you to work with images. You can read images that are interpreted as an array of `numpy`.

```python
import matplotlib.image as mpimg

grace = mpimg.imread('../Assets/grace_hopper.jpg')
print(grace.shape)
print(type(grace))


jean = mpimg.imread('../Assets/jean_sammet.jpg')
print(type(jean))

cirs = mpimg.imread('../Assets/cirs_slice.png')
```

```python
print(grace)
```

Both images have 3 channels, red, green and blue, which can be seen in the `Shape` of each of them:

```python
print(grace.shape)
print(jean.shape)
```

```python
fig5, g = plt.subplots(1, 2, figsize=[16,10])
graceplt = g[0].imshow(grace)
g[0].set_title('Grace Hopper')
plt.colorbar(graceplt,ax = g[0],orientation='horizontal')

jeanplt = g[1].imshow(jean)
g[1].set_title('Jean Sammet')
plt.colorbar(jeanplt,ax = g[1],orientation='horizontal')

plt.show()
```

```python
fig5, g = plt.subplots(1, 2, figsize=[16,10])
graceplt = g[0].imshow(grace[:,:,0], cmap = 'Paired')
g[0].set_title('Grace Hopper')
plt.colorbar(graceplt,ax = g[0],orientation='horizontal')

jeanplt = g[1].imshow(jean[:,:,0],cmap = 'Paired')
g[1].set_title('Jean Sammet')
plt.colorbar(jeanplt,ax = g[1],orientation='horizontal')

plt.show()
```

```python
fig5, g = plt.subplots(1, 2, figsize=[16,10])
g[0].imshow(grace, cmap = 'gray')
g[0].set_title('IMAGEN')

g[1].set_title('HISTOGRAMA')
g[1].hist(grace[:,:,0].ravel(), bins=100) # Ravel returns a 1D vector with all the elements of the Matrix.
g[1].set_xlim([0, 255])
g[1].set_ylim([0, 5000])

plt.show()
```

You can also calculate the histogram in `numpy`:

```python
# 
# I can normalize the image to the range 0-1
# 
img_maximo = np.max(cirs)
cirs = cirs/img_maximo


nbins = 20
c = cirs.ravel()
h = np.histogram(c,bins=nbins) # np.histogram returns a tupla with two arrays, the first
                            # It is the histogram, the second corresponds to the boundaries of the bnes
print(type(h[0]))
print(h[0])
print(h[0].shape)
print(h[1].shape)
```

But it is more laborious to do the hand histogram:
- You have to notice the difference in the size between the histogram in `h [0]` and the boundaries of the bins in `h [1]`. That is why it is necessary to select all the elements of `h [1]` except the last one with `h [1] [:-1]`. 
- We also use `align = 'edge'` so that the bars are aligned to the left of each interval.

```python
widthbar = 1/nbins*0.95
plt.bar(h[1][:-1],h[0],width = widthbar,align='edge',log=True)


print(h[0])
print(h[1][:-1])
```
